```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_1000 = pd.read_csv('datasets/dataset_1000.csv')
df_1000_hot_deck = pd.read_csv('datasets/dataset_1000_nan_Hot-deck.csv')
df_1000_mean = pd.read_csv('datasets/dataset_1000_nan_mean.csv')
df_1000_regression = pd.read_csv('datasets/dataset_1000_nan_linear_regression.csv')

df_25000 = pd.read_csv('datasets/dataset_25000.csv')
df_25000_hot_deck = pd.read_csv('datasets/dataset_25000_nan_Hot-deck.csv')
df_25000_mean = pd.read_csv('datasets/dataset_25000_nan_mean.csv')
df_25000_regression = pd.read_csv('datasets/dataset_25000_nan_linear_regression.csv')

df_100000 = pd.read_csv('datasets/dataset_100000.csv')
df_100000_hot_deck = pd.read_csv('datasets/dataset_100000_nan_Hot-deck.csv')
df_100000_mean = pd.read_csv('datasets/dataset_100000_nan_mean.csv')
df_100000_regression = pd.read_csv('datasets/dataset_100000_nan_linear_regression.csv')
```

```python
def show_distribution(df_original, df_restored):
    plt.figure(figsize=(12, 6))
    sns.kdeplot(df_original["Стоимость анализов, руб"], label="Исходные данные", color="green")
    sns.kdeplot(df_restored["Стоимость анализов, руб"], label="Восстановленные данные", color="red", linestyle="--")
    plt.title(f"Сравнение распределения 'Стоимость анализов, руб'")
    plt.legend()
    plt.show()
```

```python
def analyze_dataset(df, title):
    print(f"\n{title}\n" + "="*50)
    
    for column in df.columns:
        if pd.api.types.is_numeric_dtype(df[column]):
            mean = df[column].mean()
            median = df[column].median()
            mode = df[column].mode().iloc[0] if not df[column].mode().empty else "Нет моды"
            print(f"{column}:")
            print(f"  Среднее = {mean:.2f}")
            print(f"  Медиана = {median:.2f}")
            print(f"  Мода    = {mode}")
        else:
            mode = df[column].mode().iloc[0] if not df[column].mode().empty else "Нет моды"
            print(f"{column}:")
            print(f"  Мода    = {mode}")
```

```python
analyze_dataset(df_1000, "Исходный датасет, 1.000 строк")
```

```python
analyze_dataset(df_1000_hot_deck, "Хот-Дек после 30% удаления, 1.000 строк")

show_distribution(df_1000, df_1000_hot_deck)
```

```python
analyze_dataset(df_1000_mean, "Среднее значение после 30% удаления, 1.000 строк")

show_distribution(df_1000, df_1000_mean)
```

```python
analyze_dataset(df_1000_regression, "Линейная регрессия после 30% удаления, 1.000 строк")

show_distribution(df_1000, df_1000_regression)
```

```python
analyze_dataset(df_25000, "Исходный датасет, 25.000 строк")
```

```python
analyze_dataset(df_25000_hot_deck, "Хот-Дек после 30% удаления, 25.000 строк")

show_distribution(df_25000, df_25000_hot_deck)
```

```python
analyze_dataset(df_25000_mean, "Среднее значение после 30% удаления, 25.000 строк")

show_distribution(df_25000, df_25000_mean)
```

```python
analyze_dataset(df_25000_regression, "Линейная регрессия после 30% удаления, 25.000 строк")

show_distribution(df_25000, df_25000_regression)
```
